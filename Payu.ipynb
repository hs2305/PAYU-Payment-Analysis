# What is the average time taken between transaction & settlement (assume every day settlement happens at 12 noon)

```python
p=df[df['Settlement Date'].str.contains('21')]
```

```python
p.insert(0, 'time', ' 12:00')
```

```python
p['Settlement Date']=p['Settlement Date']+p['time']
```

```python
p
```

```python
sum=0
```

```python
import datetime
from dateutil import parser
for ind in p.index:
        s1=p['Date of Transaction'][ind]
        s2=p['Settlement Date'][ind]
        t2=parser.parse(s2)
        t1=parser.parse(s1,dayfirst='True')
        sum = sum+(((t2-t1).seconds/3600)+((t2-t1).days*24))
        
     
    
```

```python
sum
```

```python
average=sum/29981
```

```python
print(average)
```

```python
##Average time taken for a transaction=32.28 hrs
```

#  Provide hourly trend for transactions for all days together  And find out the peak/highest hours of transactions?

```python
import datetime
from dateutil import parser
for ind in df.index:
        s1=df['Date of Transaction'][ind]
        t1=parser.parse(s1,dayfirst='True')
        df['Date of Transaction'][ind]=t1
```

```python
df
```

```python
def hr_func(ts):
    return ts.hour

df['time_hour'] = df['Date of Transaction'].apply(hr_func)
```

```python
df
```

```python
sns.countplot('time_hour',data=df)
```

```python
##Peak Hour for transaction is 11:00-12:00
```

```python
p=df[['Transaction amount','time_hour']]
```

```python
p
```

```python
p.groupby('time_hour').size()
```

```python
p.groupby('time_hour')['Transaction amount'].sum()
```

##Hourly trend for transaction according to both frequency of transaction and transation amount shows that
##11:00-12:00 is peak hour of transaction
